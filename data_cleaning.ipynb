```python
# Import Libraries
import pandas as pd
import numpy as np

# Create Sample Dataset
data = {
    'Name': ['John', 'Alice', 'Bob', 'Alice', np.nan],
    'Age': [22, 25, np.nan, 25, 30],
    'City': ['New York', 'London', 'Paris', 'London', 'Tokyo'],
    'Salary': [50000, 60000, 55000, 60000, np.nan]
}

df = pd.DataFrame(data)

print("Original Dataset")
print(df)
```

```python
print("\nDataset Information:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Fill missing Age with mean value
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Fill missing Salary with mean value
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

# Fill missing Name with 'Unknown'
df['Name'] = df['Name'].fillna('Unknown')

print("\nDataset After Handling Missing Values")
print(df)
```

```python
print("\nNumber of Duplicate Rows:")
print(df.duplicated().sum())
```

```python
df = df.drop_duplicates()

print("\nDataset After Removing Duplicates")
print(df)
```

```python
df['Age'] = df['Age'].astype(int)

print("\nData Types After Conversion")
print(df.dtypes)
```

```python
print("\nStatistical Summary")
print(df.describe())
```

```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = df[(df['Salary'] < lower_limit) |
              (df['Salary'] > upper_limit)]

print("\nOutliers:")
print(outliers)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)

print("\nCleaned Dataset Saved Successfully!")
```
